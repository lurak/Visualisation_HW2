# Task 2

```python
_DAYS_IN_MONTH = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]

def _is_leap(year):
    return year % 4 == 0 and (year % 100 != 0 or year % 400 == 0)

def _days_in_month(year, month):
    if month == 2 and _is_leap(year):
        return 29
    return _DAYS_IN_MONTH[month-1]

def days(year, month, day):
    return sum([_days_in_month(int(year), m) for m in range(1, int(month))]) + int(day)
```

```python
task2_df = df.copy()
task2_df['Година'] = df['Час/Дата'].str.split('-')
task2_df['Година'] = task2_df['Година'].transform(lambda x: int(x[0]))
task2_df['День'] = df['Час/Дата'].str.split('.')
task2_df['День'] = task2_df['День'].transform(lambda x: int(x[0].split('-')[1]))
task2_df['Місяць'] = df['Час/Дата'].str.split('.')
task2_df['Місяць'] = task2_df['Місяць'].transform(lambda x: int(x[1]))
task2_df['Рік'] = df['Час/Дата'].str.split('.')
task2_df['Рік'] = task2_df['Рік'].transform(lambda x: int(x[2]))
task2_df['Day'] = task2_df.apply(lambda x : days(x['Рік'],x['Місяць'],x['День']), axis = 1)
task2_df = task2_df.groupby(['Day','Година']).sum().reset_index()
task2_df.tail(100)
```

```python
alt.Chart(task2_df).mark_rect().encode(
    x = alt.X('Година:O', title='Hours'),
    y = alt.Y('Day:O', title='Number of day'),
    color = alt.Color('CONSUMPTION:Q',title='Consumption', scale = alt.Scale(scheme = 'blueorange',domainMid=0))
).properties(title=
             alt.TitleParams('Вependence of electricity consumption on the day of the year and hour of the day'),
             height=1000)
```

- Обрав цей варіант адже хітмапа гарно працює для 3 вимірів даних
- Чесно, альтернатив в голову не приходило
- Переваги: чітко видна загальна тенденція споживання електроенергії
- Недоліки: тяжко зрозуміти що відбувалось в конкретний день

```python
task3_df = df.copy()
task3_df['hour'] = df['Час/Дата'].str.split('-')
task3_df['hour'] = task3_df['hour'].transform(lambda x: int(x[0]))
task3_df = task3_df.groupby(['hour']).sum().reset_index()
task3_df.head(30)
```

```python
new_structure = {'hour':[], 'type':[], 'production':[]}
Types = ['AES', 'TEC', 'VDE', 'TES', 'GES', 'GAES_GEN']
for index, row in task3_df.iterrows():
    for i in range(len(Types)):
        new_structure['hour'].append(row['hour'])
        new_structure['type'].append(Types[i])
        new_structure['production'].append(row[Types[i]])
task3_df = pd.DataFrame(new_structure)
task3_df.tail(15)
```

```python
alt.Chart(task3_df).mark_line().encode(
    x = alt.X('hour:Q', title='Hours'),
    y = alt.Y('production:Q', title='Production'),
    color = alt.Color('type:N',title='Types')
).properties(width = 950, height = 500, background = '#F9F9F9', padding = 25,
             title=alt.TitleParams("Change in electricity generation from different sources during the day"))
```

- Обрав цей варіант адже прості графіки з лініями легко читаються 
- Можливо зообразити це у вигляді стовпчкових діаграм
- Переваги: гарно показані тенденції кожного джерела і в цілому різниця в об'ємах генерації енергії ними
- Недоліки: джерела, що знизу трішки зливаються

```python
task4_df_1 = df.copy()
task4_df_1['hour'] = df['Час/Дата'].str.split('-')
task4_df_1['hour'] = task4_df_1['hour'].transform(lambda x: int(x[0]))
task4_df_1['month'] = df['Час/Дата'].str.split('.')
task4_df_1['month'] = task4_df_1['month'].transform(lambda x: int(x[1]))
task4_df_1 = task4_df_1.groupby(['hour', 'month']).sum().reset_index()
task4_df_1.head(10)
```

```python
alt.Chart(task4_df_1).mark_rect().encode(
    alt.X('hour:O', title='Hours', axis=alt.Axis(labelAngle=0)),
    alt.Y('month:O', title='Months'),
    alt.Color('CONSUMPTION', scale=alt.Scale(scheme='blueorange', domainMid=3000000))
).properties(height = 400,
             width=850,
             title=alt.TitleParams("Electricity consumption during the day by months"))
```

```python
def season(month):
    if 3 >= month <= 5:
        return 'Spring'
    elif 6 >= month <=8:
        return 'Summer'
    elif 9 >= month <=11:
        return 'Autumn'
    else:
        return 'Winter'
```

```python
task4_df_2 = df.copy()
task4_df_2['hour'] = df['Час/Дата'].str.split('-')
task4_df_2['hour'] = task4_df_2['hour'].transform(lambda x: int(x[0]))
task4_df_2['month'] = df['Час/Дата'].str.split('.')
task4_df_2['month'] = task4_df_2['month'].transform(lambda x: int(x[1]))
task4_df_2['season'] = task4_df_2.apply(
    lambda x : season(x['month']), axis = 1)
task4_df_2 = task4_df_2.groupby(['hour', 'season']).sum().reset_index()
task4_df_2.head()
```

```python
alt.Chart(task4_df_2).mark_rect().encode(
    alt.X('hour:O', title='Hours', axis=alt.Axis(labelAngle=0)),
    alt.Y('season:O', title='Seasons'),
    alt.Color('CONSUMPTION', scale=alt.Scale(scheme='blueorange', domainMid=3000000))
).properties(height = 400,
             width=850,
             title=alt.TitleParams("Electricity consumption during the day by seasons"))
```

```python
task5_df = df.copy()
task5_df['day'] = df['Час/Дата'].str.split('.')
task5_df['day'] = task5_df['day'].transform(lambda x: int(x[0].split('-')[1]))
task5_df['month'] = df['Час/Дата'].str.split('.')
task5_df['month'] = task5_df['month'].transform(lambda x: int(x[1]))
task5_df['year'] = df['Час/Дата'].str.split('.')
task5_df['year'] = task5_df['year'].transform(lambda x: int(x[2]))
task5_df.head()
```

```python
def day_of_week(year, mth, day):
    day = datetime.datetime(year, mth, day)
    return day.weekday()
task5_df['Day_of_week'] = task5_df.apply(
    lambda x : day_of_week(x['year'],x['month'],x['day']), axis = 1)
task5_df = task5_df.groupby(['Day_of_week']).sum().reset_index()
task5_df.head(7)
```

```python
alt.Chart(task5_df).mark_bar().encode(
                     x = alt.X('Day_of_week:O', title='Day of the weeek',axis=alt.Axis(labelAngle=0)),
                     y=alt.Y('CONSUMPTION', title='Consumption',scale = alt.Scale(
                  zero = False))
                     ).properties(
                                  height = 400,
                                  width=850,
     title=alt.TitleParams("Electricity consumption during the week")
)
```
